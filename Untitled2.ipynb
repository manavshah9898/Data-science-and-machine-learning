```
import queue
import time

x = 0  # jug 1
y = 0  # jug 2
MaxJug1 = 4  # max capacity of jug 1
MaxJug2 = 3  # max capacity of jug 2

print("Initial state = (0,0)")
print("Capacities = (4,3)")

BFS_Q = queue.


class Node:
    # initializing a constructor for creating a node
    def __init__(self, parent):
        self.x = 0
        self.y = 0
        self.parent = parent

    def isGoalNode(self, GNode):
        if self.x == GNode:
            return True
        return False

    def printNode(self):
        print("(", self.x, ",", self.y, ")")


def generateAllSuccessor(cnode):
    list1 = []
    for Rule in range(1, 9):
        next_node = operation(cnode, Rule)  # current node
        if next_node is not None:
            list1.append(next_node)  # next node
    return list1


# checking a rule
def operation(cnode, Rule):
    x = cnode.x
    y = cnode.y

    if Rule == 1:
        if x < MaxJug1:
            x = MaxJug1
        else:
            return None
    elif Rule == 2:
        if y < MaxJug2:
            y = MaxJug2
        else:
            return None
    elif Rule == 3:
        if x > 0:
            x = 0
        else:
            return None
    elif Rule == 4:
        if y > 0:
            y = 0
        else:
            return None
    elif Rule == 5:
        if x + y >= MaxJug1:
            y = y - (MaxJug1 - x)
            x = MaxJug1
        else:
            return None
    elif Rule == 6:
        if x + y >= MaxJug2:
            x = x - (MaxJug2 - y)
            y = MaxJug2
        else:
            return None
    elif Rule == 7:
        if x + y < MaxJug1:
            x = x + y
            y = 0
        else:
            return None
    elif Rule == 8:
        if x + y < MaxJug2:
            y = x + y
            x = 0
        else:
            return None
    else:
        print("invalid input!")

    nextnode = Node(cnode)
    nextnode.x = x
    nextnode.y = y
    nextnode.parent = cnode
    return nextnode


def PushList(list1):
    for m in list1:
        BFS_Q.put(m)


def PopNode():
    if (BFS_Q.empty()):
        return None
    else:
        return BFS_Q.get()


def bfs(InitialNode, GNode):
    BFS_Q.put(InitialNode)
    while not BFS_Q.empty():
        visitednode = PopNode()
        if visitednode.isGoalNode(GNode):
            return visitednode
        successor_node = generateAllSuccessor(visitednode)
        PushList(successor_node)
    return None


def PrintPath(CNode):
    temp = CNode
    list2 = []
    while temp is not None:
        list2.append(temp)
        temp = temp.parent
    list2.reverse()
    for i in list2:
        i.printNode()
    print("path:", len(list2))


initial_node = Node(None)
currentNode = initial_node
GNode = int(input("enter goal state of jug1:"))
if GNode <= MaxJug1:
    GNode = GNode
else:
    print("capacity must be less than maximum capacity of jug 1!")
    exit()
print("Goal state = ({},y)".format(GNode))
var1 = int(input("enter 1 for manually and enter 2 for bfs:"))

if var1 == 1:
    while not currentNode.isGoalNode(GNode):
        rule = int(input("enter rule no.:"))
        NextNode = operation(currentNode, rule)
        if NextNode is not None:
            currentNode = NextNode
            currentNode.printNode()

elif var1 == 2:
    start_time = time.time()
    solution = bfs(initial_node, GNode)
    end_time = time.time()
    if solution is None:
        print("solution not found")
    else:
        PrintPath(solution)
    print("execution time", round((end_time - start_time) * 1000, 2), " millis")

else:5
8
    print("invalid input")
```
